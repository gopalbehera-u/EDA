# piechart

```python
plt.pie(
    dataset['Embarked'].value_counts(),
    labels=dataset['Embarked'].value_counts().index,
    autopct='%.2f%%'
)

plt.show()
```

```python
plt.pie(
    x=dataset['Pclass'].value_counts(),
    labels=dataset['Pclass'].value_counts().index,
    autopct='%.2f%%'
)
plt.show()
```

```python
plt.pie(
    x=dataset['Sex'].value_counts(),
    labels=dataset['Sex'].value_counts().index,
    autopct="%.2f%%"
)
```

```python
plt.pie(
    dataset['Survived'].value_counts(),
    labels=dataset['Survived'].value_counts().index,
    autopct="%.2f%%"
)
plt.show()
```
